```python
# Passo 1: Importar as bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ferramentas do Scikit-Learn para Machine Learning
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

# Passo 2: Simular a leitura de um arquivo CSV (substitua pelo seu arquivo real)
# Criando um dataset fictício
dados_ficticios = {
    'area_m2': [50, 70, 85, 120, 45, 150, 90, 110, 65, 130],
    'quartos': [1, 2, 3, 3, 1, 4, 2, 3, 2, 4],
    'vagas_garagem': [0, 1, 1, 2, 0, 3, 1, 2, 1, 2],
    'preco_venda': [220000, 310000, 390000, 550000, 195000, 720000, 410000, 510000, 290000, 600000]
}

# Transformando em um DataFrame do Pandas
df = pd.DataFrame(dados_ficticios)
print("--- Primeiras linhas do seu conjunto de dados ---")
print(df.head(), "\n")

# Passo 3: Definir as variáveis x e y
# x = Tudo o que o modelo vai usar para tentar adivinhar o preço (características)
# y = O que o modelo precisa adivinhar (preço final)
x = df[['area_m2', 'quartos', 'vagas_garagem']]
y = df['preco_venda']

# Passo 4: Dividir os dados em treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.20, random_state=42)

print("--- Tamanho das divisões ---")
print(f"Dados totais: {x.shape[0]} imóveis")
print(f"Treino (x_treino): {x_treino.shape[0]} imóveis -> Modelo vai estudar aqui")
print(f"Teste (x_teste): {x_teste.shape[0]} imóveis -> Aqui vamos testar a prova final")

# Passo 5: Instanciar e treinar o modelo 
modelo = LinearRegression()
modelo.fit(x_treino, y_treino)

print("🎉 Modelo treinado com sucesso!\n")

# Passo 6: Fazer previsões e avaliar o modelo
previsoes = modelo.predict(x_teste)

# Calculando o MAE (Erro Médio Absoluto)
mae = mean_absolute_error(y_teste, previsoes)

# Calculando o R² (Coeficiente de Determinação)
r2 = r2_score(y_teste, previsoes)

print("--- Avaliação do Modelo ---")
print(f"MAE (Erro Médio Absoluto): R$ {mae:,.2f}")
print(f"R² (Score de Precisão): {r2:.2f}")

# Passo 7: Comparando na prática: Preço Real vs. Previsão do Modelo
comparacao = pd.DataFrame({
    'Preço Real': y_teste.values,
    'Previsão do Modelo': previsoes
})
comparacao['Diferença (R$)'] = abs(comparacao['Preço Real'] - comparacao['Previsão do Modelo'])

print("\n--- Comparação Real vs. Previsão nos dados de Teste ---")
print(comparacao)
```
